```python
from skimage.color import rgb2gray
from skimage import io
import matplotlib.pyplot as plt
```

```python
class Edit:
    i = 0
    def horizontal_flip(self,np_image):
        image_flip = np_image[:,::-1]      # horizontal flip
        plt.imshow(image_flip)
        return image_flip
        
        
    def vertical_flip(self,np_image):
        image_flip = np_image[::-1]        # vertical flip
        plt.imshow(image_flip)
        return image_flip
        
    def invert_colors(self,np_image):           # inverting colors
        image_invert = 255 - np_image
        plt.imshow(image_invert)
        return image_invert
        
    def convert_to_grayscale(self,np_image):         # gray scaling
        image_gray_scale = rgb2gray(np_image)
        plt.imshow(image_gray_scale,cmap = 'gray',vmin= 0, vmax= 1)
        return image_gray_scale
        
    
    def convert_to_black_white(self,np_image):          # black and white
        image_black_white = rgb2gray(np_image)
        image_black_white[image_black_white >= 0.85] = 1
        image_black_white[image_black_white < 0.85] = 0
        plt.imshow(image_black_white,cmap = 'gray', vmin = 0,vmax = 1)
        return image_black_white
    
```

```python
class Solution:
    def __init__(self):
        self.__np_image1 = None
        self.new_image = None
        
        print("-----------Welcome to Basic Image Processing using Numpy-----------")
        while True:
            print("Select 1: to load an image")
            print("Select 2: to edit an image")
            print("Select 3: to save an image")
            print("Select 4: to quit")
            print()

            e = Edit()                  # creating an object of Edit class

            inp = int(input("Please enter your option: "))

            if inp == 1:
                '''Loading a file'''
                try:
                    file = input("Please enter your file name here: ") 
                    self.__np_image1 = io.imread(file)
                    print("File loaded!")
                    print()
                except:
                    print("File not Found!")


            elif inp == 2:
                Edit.i += 1
                '''Editing an image'''
                print("Select 21: horizontal flip")
                print("Select 22: vertical flip")
                print("Select 23: invert colors")
                print("Select 24: convert to Grayscale")
                print("Select 25: convert to Black&White")

                op = int(input("Select your edit option: "))

                if op == 21:
                    '''Horizontal flip'''
                    self.new_image = e.horizontal_flip(self.__np_image1)
                    print("Horizontal flip is done!")
                    print()

                elif op == 22:
                    '''Vertical flip'''
                    self.new_image = e.vertical_flip(self.__np_image1)
                    print("Vertical flip is done!")
                    print()

                elif op == 23:
                    '''Invert Colors'''
                    self.new_image = e.invert_colors(self.__np_image1)
                    print("Inverting Colors is done!")
                    print()

                elif op == 24:
                    '''Convert to GrayScale'''
                    self.new_image = e.convert_to_grayscale(self.__np_image1)
                    print("Conversion to GrayScale is done!")
                    print()

                elif op == 25:
                    '''Convert to Black-White'''
                    self.new_image = e.convert_to_black_white(self.__np_image1)
                    print("Conversion to Black-White is done!")
                    print()

                else:
                    print("Invalid option")
                    break
                    exit(1)




            elif inp == 3:
                '''Save Image'''
                try:
                    io.imsave(fname = 'D:\Pictures\Edit\edited_image{}.jpg'.format(Edit.i),arr = self.new_image)
                    print("Image is successfully saved!")
                    print()
                    
                except:
                    print("Error occured while saving the image")

            elif inp == 4:
                '''Quit'''
                break
                exit(0)

            else:
                print("You entered an invalid choice.")
                break
                exit(1)
            
            
            
        
```
