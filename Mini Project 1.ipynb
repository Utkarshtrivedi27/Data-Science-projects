```python
class Account:
    def __init__(self):
        self._pin = int(input("Please set your pin: "))     # pin is set as protected and balance is set as private variable for the notion of encapsulation.
        self.__balance = int(input("Enter the amount to credit your account: "))
        print()
    
    def withdraw(self,amount):
        '''updates the account balance on withdraw'''
        self.__balance -= amount
        
    def deposit(self,amount):
        '''updates the account balance on deposit'''
        self.__balance += amount
        
    def transfer(self,to_acc2,amount):
        '''debites the savings account with a transfer amount and credits the demat account with that same amount.
        It returns 0 if the transfer is failed, otherwise returns 1.'''
        if amount > self.__balance:
            print("Transfer failed due to insufficient account balance.")
            return 0
            
        else:
            self.__balance -= amount
            to_acc2.__balance += amount
            return 1
        
    def show_balance(self):
        '''returns the account balance'''
        return self.__balance
    
    
    
    
    
    
class ATM:
    def __init__(self,acc1,acc2):
        again = ""
        choice = '1'
        inp = 0
        while True:
            if choice == '0':
                break
            if inp not in (0,1,2,3,4):
                break

            pin = int(input("Please enter your Savings Account pin {}: ".format(again)))
            if pin == acc1._pin:
                print("----------------Welcome-----------------")
                while True:
                    print("Select 1 : To show your account balance")
                    print("Select 2 : To deposit")
                    print("Select 3 : To withdraw")
                    print("Select 4 : To transfer")
                    inp = int(input("Please select [1-4]: "))
                    if inp == 1:
                        print("Your Savings account balance is {}.".format(acc1.show_balance()) )
                        
                    elif inp == 2:
                        deposit_amt = int(input("Enter your deposit amount: "))
                        acc1.deposit(deposit_amt)
                        print("Your Savings account is credited with {} and your Savings account balance is {}.".
                              format(deposit_amt, acc1.show_balance()) )

                    elif inp == 3:
                        withdraw_amt = int(input("Enter your amount to be withdrawn: "))
                        acc1.withdraw(withdraw_amt)
                        print("Your Savings account is debited with {} and your Savings account balance is {}.".
                              format(withdraw_amt, acc1.show_balance()) )

                    elif inp == 4:
                        acc_no = acc2
                        transfer_amt = int(input("Enter your amount to be transfered: "))
                        status = acc1.transfer(acc_no,transfer_amt)
                        
                        if status == 1:
                            print("Your Savings account is debited with {} and your Savings account balance is {}.".
                                  format(transfer_amt, acc1.show_balance()) )
                            print("Your Savings account is debited with {} and your Savings account balance is {}.".
                                  format(transfer_amt, acc2.show_balance()) )
                        
                    else:
                        print("Invalid choice")
                        break
                        

                    choice = input("Do want to exit[< 0 > for Yes/ <any number> for No]?: ")
                    print()
                    
                    if choice == '0':
                        print("Thanks for using our services")
                        break

            else:
                print("You entered the wrong pin.\n")
                again = "again"
```
